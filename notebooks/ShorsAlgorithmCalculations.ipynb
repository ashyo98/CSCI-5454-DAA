```python
from matplotlib import pyplot as plt
```

```python
N = 13 * 7
```

```python
def find_period(r, N):
    prod = r
    for j in range(2, N):
        prod = (prod * r) % N
        if prod == 1:
            print(f'Period = {j}')
            return j
    return None
```

```python
def plot_period(r, N, nperiods=4):
    lst = [r]
    prod = r
    num_times = 0
    for j in range(2,N):
        prod = (prod * r) % N
        lst.append(prod)
        if prod == 1: # 1 period complete
            num_times = num_times + 1
            if num_times >= nperiods:
                return lst 
    return lst
```

```python
r = 15
find_period(r, N)
plt.stem(plot_period(r, N, 6), '--x')
```

```python
r = 19
N = 91
period = find_period(r, N)
plt.stem(plot_period(r, N, 6), '--x')
```

```python
def gcd(a, b):
    (a,b) = (max(a,b), min(a,b))
    
    assert b != 0
    
    if a == b:
        return a
    
    if b == 1:
        return a
    
    while (b > 1):
        (a, b) = (b,  a% b)
    return a
```

```python
def factor_with_period(r, period, N):
    assert period % 2 == 0
    t = period // 2
    prod = r
    j = 0
    result = 1
    while t > 0:
        if t %2 == 1: 
            result = (result * prod) % N
        prod = (prod * prod) % N
        t = t//2
        print(t, prod, result)
    p1 = gcd(result-1, N)
    p2 = gcd(result+1, N)
    return (p1, p2)
```

```python
from scipy.fft import fft, ifft
```

```python
periodic = [1 if (j-1) % 5 == 0 else 0 for j in range(15) ]
plt.stem(periodic)
```

```python
fourier = fft(periodic)
amplitudes = [(z.real * z.real + z.imag * z.imag) for z in fourier]
plt.stem(amplitudes)
```
